#  Question 2

### i) Determine from the dataset what percentage of all recorded shark attacks were fatal.

```python
wdf['Fatal'].value_counts()
```

```python
num_rows = len(wdf.index)
num_fatal = wdf['Fatal'].value_counts().tolist()[1]
result = num_fatal * 100 / num_rows
print(f'Percentage of attacks that are:\n{round(result, 2)}')
```

### ii) For each individual country print out the percentage of shark attacks.

```python
# To know what columns needed

wdf.columns
```

```python
# Display countries names

wdf['Country'].value_counts().keys()
```

```python
# create a list of countries

country_list = wdf['Country'].value_counts().keys()


for country in country_list:
    fatal = wdf[(wdf['Country'] == country) & (wdf['Fatal'] == 'Y')]
    non_fatal = wdf[(wdf['Country'] == country) & (wdf['Fatal'] == 'N')]
    if len(fatal) > 0:
        result = len(fatal) * 100 / (len(fatal) + len(non_fatal))
        print(f'The percentage of fatal attacks: {country} {round(result, 2)}%')
    else:
        print(f'No fatal attacks in: {country}')
```

#  Question 3

### i) In this question we are interested in looking at the number of recorded shark attacks over time for a specific country.


```python
wdf.columns
```

```python
def calculate_yearly_attacks(country, wdf):
    years_list = pd.unique(wdf['Year'][(wdf['Year'] >= 1925) & (wdf['Year'] <= 2015)].to_list())

    for year in years_list:
        result = len(wdf[(wdf['Country'] == country) & (wdf['Year'] == year)])
        print(f'Number of attacks in {country} during {int(year)} is {result}\n------')

calculate_yearly_attacks('AUSTRALIA', wdf)
```
